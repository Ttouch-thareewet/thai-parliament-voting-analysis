```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("final.csv")
```

```python
df
```

```python
df_selected = df[["vote_event_id","vote_event_date", "vote_event_result", "vote_event_classification", "person_name", "voter_party","option","pm_name","is_prayut"]]
```

```python
df_selected["voter_party"].value_counts().to_csv("party_counts.csv")
```

```python

party_counts = df_selected["voter_party"].value_counts()


valid_parties = party_counts[
    (party_counts >= 700) & (party_counts.index != "สมาชิกวุฒิสภา")
].index


df_filtered = df_selected[df_selected["voter_party"].isin(valid_parties)]
```

```python
df_filtered["option"].value_counts()
```

```python
valid_votes = [
    "เห็นด้วย",
    "ไม่เห็นด้วย",
    "งดออกเสียง",
    "ไม่ลงคะแนนเสียง",
    "ลา / ขาดลงมติ"
]

df_filtered = df_filtered[df_filtered["option"].isin(valid_votes)]
```

```python
df_filtered["option"].value_counts()
```

```python
vote_map = {
    "เห็นด้วย": 1,
    "ไม่เห็นด้วย": -1,
    "งดออกเสียง": 0,
    "ไม่ลงคะแนนเสียง": 0,
    "ลา / ขาดลงมติ": 0
}

df_filtered["vote_num"] = df_filtered["option"].map(vote_map)
```

```python
df_filtered
```

```python
df_filtered["vote_event_date"] = pd.to_datetime(df_filtered["vote_event_date"])
cut_date = "2023-05-05"

df_before = df_filtered[df_filtered["vote_event_date"] < cut_date]
df_after = df_filtered[df_filtered["vote_event_date"] >= cut_date]
```

```python
print(len(df_before))
print(len(df_after))
```

```python
df_before
```

```python
vote_matrix_before = df_before.pivot_table(
    index="person_name",
    columns="vote_event_id",
    values="vote_num"
)
```

```python
vote_matrix_before
```

```python
vote_counts = vote_matrix_before.notna().sum(axis=1)
threshold = 0.3 * vote_matrix_before.shape[1]

valid_mps = vote_counts[vote_counts >= threshold].index
```

```python
vote_matrix_before = vote_matrix_before.loc[valid_mps]
```

```python
vote_matrix_before = vote_matrix_before.fillna(0)
```

```python
vote_matrix_before
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = StandardScaler().fit_transform(vote_matrix_before)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
print(pca.explained_variance_ratio_)
print("Total variance:", pca.explained_variance_ratio_.sum())
```

```python
pca_df = pd.DataFrame(
    X_pca,
    columns=["PC1", "PC2"],
    index=vote_matrix_before.index
)
party_label = df_before.groupby("person_name")["voter_party"].first()

pca_df["party"] = party_label.loc[pca_df.index]
```

```python
pca_df["party"].value_counts()
```

```python
party_colors_before = {
    "เพื่อไทย": "red",
    "พลังประชารัฐ": "skyblue",
    "ภูมิใจไทย": "purple",
    "ก้าวไกล": "orange",
    "ประชาธิปัตย์": "green",
    "ชาติไทยพัฒนา": "brown",
    "เสรีรวมไทย": "pink",
    "ประชาชาติ": "gold"
}
```

```python
pca_before_df = pca_df.copy()
```

```python

plt.rcParams["font.family"] = "Tahoma"


plt.figure(figsize=(8,6))

sns.scatterplot(
    data=pca_before_df,
    x="PC1",
    y="PC2",
    hue="party",
    palette=party_colors_before,
    s=120,
    alpha=0.7
)

plt.axhline(0, linestyle="--", color="gray")
plt.axvline(0, linestyle="--", color="gray")

plt.title("PCA การโหวตของ ส.ส. (ก่อน 2023-05-05)")

plt.legend(bbox_to_anchor=(1.05,1), loc="upper left")

plt.show()
```

```python
df_after.loc[
    df_after["voter_party"] == "ก้าวไกล",
    "voter_party"
] = "ประชาชน"
```

```python
vote_matrix_after = df_after.pivot_table(
    index="person_name",
    columns="vote_event_id",
    values="vote_num"
)
```

```python
# vote_matrix_after.loc["เฉลิม อยู่บำรุง"].to_csv("เฉลิม.csv")
```

```python
vote_matrix_after
```

```python
vote_counts = vote_matrix_after.notna().sum(axis=1)

threshold = 0.3 * vote_matrix_after.shape[1]

valid_mps = vote_counts[vote_counts >= threshold].index

vote_matrix_after = vote_matrix_after.loc[valid_mps]
```

```python
vote_matrix_after = vote_matrix_after.fillna(0)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = StandardScaler().fit_transform(vote_matrix_after)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print(pca.explained_variance_ratio_)
print("Total variance:", pca.explained_variance_ratio_.sum())
```

```python
pca_df = pd.DataFrame(
    X_pca,
    columns=["PC1","PC2"],
    index=vote_matrix_after.index
)
```

```python
party_label = df_after.groupby("person_name")["voter_party"].first()

pca_df["party"] = party_label.loc[pca_df.index]
```

```python
pca_df["party"].value_counts()
```

```python
party_colors_after = {
    "เพื่อไทย": "#E31A1C",        # แดง
    "ประชาชน": "#FF7F00",         # ส้ม
    "ภูมิใจไทย": "#6A3D9A",       # ม่วง
    "พลังประชารัฐ": "#1F78B4",    # ฟ้า
    "รวมไทยสร้างชาติ": "#0033A0", # น้ำเงินเข้ม
    "ประชาธิปัตย์": "#33A02C",    # เขียว
    "ชาติไทยพัฒนา": "#8C564B",   # น้ำตาล
    "ประชาชาติ": "#B15928",      # น้ำตาลแดง
    "กล้าธรรม": "#17BECF",       # turquoise
    "เสรีรวมไทย": "#F781BF"      # ชมพู
}
```

```python
pca_after_df = pca_df.copy()
```

```python

plt.rcParams["font.family"] = "Tahoma"

plt.figure(figsize=(9,7))

sns.scatterplot(
    data=pca_after_df,
    x="PC1",
    y="PC2",
    hue="party",
    palette=party_colors_after,
    s=120,
    alpha=0.75,
    edgecolor="black"
)


plt.axhline(0, linestyle="--", color="gray", linewidth=1)
plt.axvline(0, linestyle="--", color="gray", linewidth=1)

plt.title("PCA การโหวตของ ส.ส. (หลัง 2023-05-05)", fontsize=14)


plt.legend(
    title="พรรค",
    bbox_to_anchor=(1.02,1),
    loc="upper left"
)

plt.tight_layout()
plt.show()
```
